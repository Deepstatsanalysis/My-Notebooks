```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

%matplotlib inline
```

```python
train_values = pd.read_csv('train_values.csv')
train_labels = pd.read_csv('train_labels.csv')

train_data = train_values
train_data['accepted'] = train_labels['accepted']
train_data.shape
```

```python
cols = ['state_code', 'county_code', 'msa_md']
for col in cols:
    train_data.loc[train_data[col] == -1, col] = np.nan
```

```python
train_data.isnull().sum()
```

```python
# column_medians = [train_data[col].median() for col in train_data.columns]
# Fill missing values with column median
for col in train_data.columns:
    train_data.loc[train_data[col].isnull(), col] = train_data[col].median()
```

```python
train_data.isnull().sum()
```

```python
train_data.to_csv('preped_data2.csv')
```
